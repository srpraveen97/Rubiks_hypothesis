```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
p_data = pd.read_csv("P_data.csv")
n_data = pd.read_csv("N_data.csv")
p_data = p_data.rename(columns={"Best_time":"Best_Time (seconds)"})
n_data = n_data.rename(columns={"Best_time":"Best_Time (seconds)"})
n_data = n_data.drop(['Day'],axis=1)
```

```python
sns_n = sns.histplot( n_data['Best_Time (seconds)'] , color="red", label="Brothers_time", kde=True);
fig = sns_n.get_figure()
fig.savefig("Naveen_dist.png")
```

```python
fig = plt.figure()
stats.probplot(n_data['Best_Time (seconds)'], plot=sns.mpl.pyplot);
fig.savefig("Naveen_qq.png")
```

```python
sns_n = sns.histplot(p_data['Best_Time (seconds)'] , color="skyblue", label="Brothers_time", kde=True);
fig = sns_n.get_figure()
fig.savefig("Praveen_dist.png")
```

```python
fig = plt.figure()
stats.probplot(p_data['Best_Time (seconds)'], plot=sns.mpl.pyplot);
fig.savefig("Praveen_qq.png")
```

```python
fig = plt.figure()
sns.histplot( p_data['Best_Time (seconds)'] , color="skyblue", label="My_time",kde=True)
sns.histplot( n_data['Best_Time (seconds)'] , color="red", label="Brothers_time",kde=True)
plt.legend();
fig.savefig("Combined.png")
```
